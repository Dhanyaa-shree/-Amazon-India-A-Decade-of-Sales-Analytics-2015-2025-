###Data Inconsistency Handling

####1)Standardize Boolean Columns

```python
bool_cols = ['is_prime_member', 'is_prime_eligible', 'is_festival_sale']

for col in bool_cols:
    df[col] = (
        df[col].astype(str).str.strip().str.lower()
        .replace({
            'true': True, 'false': False,
            '1': True, '0': False,
            'yes': True, 'no': False
        })
    )
```

####2)Fix Category Column (case + spelling issues)

```python
df['category'] = df['category'].str.strip().str.lower()

df['category'] = df['category'].replace({
    'electronic': 'electronics',
    'electronicss': 'electronics',
    'electronics & accessories': 'electronics'
})
```

####3)Clean Customer City Names

```python
df['customer_city'] = df['customer_city'].str.strip().str.lower()

city_map = {
    'bombay': 'mumbai',
    'mumba': 'mumbai',
    'bangalore': 'bengaluru',
    'banglore': 'bengaluru',
    'calcutta': 'kolkata',
    'madras': 'chennai',
    'new delhi': 'delhi',
    'delhi ncr': 'delhi'
}

df['customer_city'] = df['customer_city'].replace(city_map).str.title()
```

####4)Fix Customer Rating Format

```python
import re

df['customer_rating'] = (
    df['customer_rating']
    .astype(str)
    .str.extract(r'(\d+\.?\d*)')[0]
)

df['customer_rating'] = pd.to_numeric(df['customer_rating'], errors='coerce')
```

####5)Clean Delivery Days

```python
import numpy as np
import pandas as pd
import re

def clean_delivery(x):
    if pd.isna(x):
        return np.nan

    x = str(x).strip().lower()

    # empty string check
    if x == "" or x == "nan":
        return np.nan

    # special cases
    if "same" in x or "express" in x:
        return 0

    # range case (1-2 days)
    match = re.match(r'(\d+)\s*-\s*(\d+)', x)
    if match:
        a, b = match.groups()
        return (float(a) + float(b)) / 2

    # extract single number
    num = re.search(r'\d+', x)
    if num:
        return float(num.group())

    return np.nan

df['delivery_days'] = df['delivery_days'].apply(clean_delivery)
```

###Outlier Detection & Handling

```python
numeric_cols = [
    'original_price_inr',
    'discounted_price_inr',
    'final_amount_inr',
    'quantity',
    'delivery_days',
    'product_weight_kg',
    'customer_rating'
]

df[numeric_cols].skew().sort_values(ascending=False)
```

```python
numeric_cols = [
    'original_price_inr',
    'discounted_price_inr',
    'final_amount_inr',
    'quantity',
    'delivery_days',
    'product_weight_kg',
    'customer_rating'
]

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = ((df[col] < lower) | (df[col] > upper)).sum()

    print(f"{col}: {outliers} outliers")
```

```python
def flag_outliers_iqr(df, col):
    outlier_mask = pd.Series(False, index=df.index)

    for (sub, br), group in df.groupby(['subcategory', 'brand']):
        q1 = group[col].quantile(0.25)
        q3 = group[col].quantile(0.75)

        iqr = q3 - q1

        lower = q1 - 1.5 * iqr
        upper = q3 + 1.5 * iqr

        outlier_mask.loc[group.index] = (
            (group[col] < lower) |
            (group[col] > upper)
        )

    return outlier_mask

df['price_outlier_IQR'] = flag_outliers_iqr(df, 'original_price_inr')
```

```python
def flag_outliers_sigma(df, col):
    outlier_mask = pd.Series(False, index=df.index)

    for (sub, br), group in df.groupby(['subcategory', 'brand']):

        mean = group[col].mean()
        std = group[col].std()

        if std == 0 or pd.isna(std):
            continue

        lower = mean - 3 * std
        upper = mean + 3 * std

        outlier_mask.loc[group.index] = (
            (group[col] < lower) |
            (group[col] > upper)
        )

    return outlier_mask

df['price_outlier_3sigma'] = flag_outliers_sigma(df, 'original_price_inr')
```

```python
print("IQR Outliers:",
      df['price_outlier_IQR'].sum())

print("3-Sigma Outliers:",
      df['price_outlier_3sigma'].sum())
```
